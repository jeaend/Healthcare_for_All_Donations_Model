```python
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder

# Separate numerical and categorical features
numerical_features = X.select_dtypes(include=['number'])
categorical_features = X.select_dtypes(include=['object'])

# MinMax scale the numerical features
scaler = MinMaxScaler()
scaled_numerical_features = pd.DataFrame(scaler.fit_transform(numerical_features), columns=numerical_features.columns)

# Ordinal encode the categorical features
encoder = OrdinalEncoder()
encoded_categorical_features = pd.DataFrame(encoder.fit_transform(categorical_features), columns=categorical_features.columns)

# Concatenate scaled numerical and encoded categorical features
X = pd.concat([scaled_numerical_features, encoded_categorical_features], axis=1)
```

```python
X
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier()
rf_classifier.fit(X, y)
```

```python
# Make predictions on all data
y_pred_all = rf_classifier.predict(X)

# Optionally, you can append the predictions to your DataFrame
X['predicted_target'] = y_pred_all

# Print or inspect the predictions
print(X['predicted_target'])
```

```python
X['actual_target'] = y
X
```

```python
# Calculate the number of correct predictions
num_correct_predictions = (X['predicted_target'] == X['actual_target']).sum()

# Calculate the total number of predictions
total_predictions = len(X)

# Calculate the accuracy
accuracy = num_correct_predictions / total_predictions
accuracy
```
